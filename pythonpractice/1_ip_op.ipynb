## Primitive Datatypes

- float
- str

## Collection Datatypes

```python
a = 10
b = 10.5
c = "Python"
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

# Variables

- Variables are used to store data
- No datatype declaration required in Python

```python
x = 10
name = "Keerthi"
price = 99.99

print(x)
print(name)
print(price)
```

```python
# Dynamic typing

a = 10
print(a)

a = "Python"
print(a)
```

# Global Variable

```python
x = 10

def update():
    global x
    x = 20

update()

print(x)
```

# Variable Naming Styles

- camelCase
- PascalCase
- snake_case

```python
studentName = "Ram"
StudentName = "Ram"
student_name = "Ram"

print(studentName)
print(StudentName)
print(student_name)
```

# Input

```python
# Static Input
a, b, c = 1, "hero", "Ram Charan"

print(a)
print(b)
print(c)
```

```python
# Single value to multiple variables
x = y = z = 100

print(x)
print(y)
print(z)
```

```python
# Dynamic Input
name = input("Enter your name: ")
age = int(input("Enter your age: "))
height = float(input("Enter your height: "))

print(name)
print(age)
print(height)
```

# Output

```python
name = "Keerthi"
age = 21
height = 5.45689

print(f"Name: {name}\nAge: {age}\nHeight: {height:.2f}")
```
