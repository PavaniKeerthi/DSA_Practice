# Conditional Statements

## Types
- if
- if else
- if elif else

## Short Hand If-Else
- Ternary Operator

```python
# if statement

age = 18

if age >= 18:
    print("Eligible to vote")
```

```python
# if else statement

number = 10

if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# if elif else statement

marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

# Single-line Format

```python
age = 20

if age >= 18: print("Adult")
```

```python
number = 5

if number > 0: print("Positive")
else: print("Non-positive")
```

# Short Hand If-Else

```python
number = 10
#syntax: <expression1> if <condition> else <expression2>
result = "Positive" if number > 0 else "Non-positive"

print(result)
```

```python
a = 20
b = 15

largest = a if a > b else b

print(largest)
```
