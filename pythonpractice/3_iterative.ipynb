# Iterative Statements

## Types
- for loop
- while loop

## Loop Controls
- break
- continue
- pass

```python
# for loop

for i in range(5):
    print(i)
```

```python
# while loop

i = 0

while i < 5:
    print(i)
    i += 1
```

# Iterating Through Lists

```python
numbers = [10, 20, 30, 40]

# Index-Based Iteration
for i in range(len(numbers)):
    print(i, numbers[i])
```

```python
numbers = [10, 20, 30, 40]

# Element-Based Iteration
for value in numbers:
    print(value)
```

```python
numbers = [10, 20, 30, 40]

# Index + Element Iteration
for i, value in enumerate(numbers):
    print(i, value)
```

```python
numbers = [10, 20, 30, 40]

# while loop Iteration
i = 0

while i < len(numbers):
    print(i, numbers[i])
    i += 1
```

# Loop Control Statements

```python
# break

for i in range(5):

    if i == 3:
        print("break executed")
        break

    print(i)
```

```python
# continue

for i in range(5):

    if i == 2:
        print("continue executed")
        continue

    print(i)
```

```python
# pass

for i in range(5):

    if i == 3:
        print("pass executed")
        pass

    print(i)
```
