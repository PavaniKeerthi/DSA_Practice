# Set

## Initialization
- set()
- Unique elements only
- Unordered collection

```python
# Empty set
s = set()

print(s)
```

```python
# Set with values
s = {1, 2, 3, 4}

print(s)
```

```python
# Duplicate values removed automatically
s = {1, 2, 2, 3, 3, 4}

print(s)
```

# Insert Operations

```python
s = {1, 2, 3}

# add()
s.add(4)

print(s)
```

```python
s = {1, 2}

# update()
s.update([3, 4, 5])

print(s)
```

# Delete Operations

```python
s = {1, 2, 3, 4}

# Remove element
new_set = s - {2}   #set operation

print(new_set)
```

```python
s = {1, 2, 3}

# clear()
s.clear()

print(s)
```

# Update Elements

```python
s = {1, 2, 3}

# Remove old value and add new value
s.remove(2)
s.add(20)

print(s)
```

# Sorting & Reverse

```python
s = {5, 2, 1, 4}

# sorted()
new_list = sorted(s)

print(new_list)
```

```python
s = {1, 2, 3, 4, 5}

# Reverse
rev_list = list(s)[::-1]

print(rev_list)
```

# Accessing Elements

```python
s = {10, 20, 30}

# Looping
for item in s:
    print(item)
```

# Set Comprehension

```python
numbers = [1, 2, 3, 4, 5, 6]

squares = {x**2 for x in numbers}

print(squares)
```

```python
numbers = [1, 2, 3, 4, 5, 6]

even_squares = {x**2 for x in numbers if x % 2 == 0}

print(even_squares)
```

# Searching

```python
s = {10, 20, 30}

# in operator
print(20 in s)
print(100 in s)
```

# Set Operations

```python
A = {1, 2, 3}
B = {3, 4, 5}

# Union
print(A | B)
print(A.union(B))
```

```python
A = {1, 2, 3}
B = {3, 4, 5}

# Intersection
print(A & B)
print(A.intersection(B))
```

```python
A = {1, 2, 3}
B = {3, 4, 5}

# Difference
print(A - B)
print(A.difference(B))
```

```python
A = {1, 2, 3}
B = {3, 4, 5}

# Symmetric Difference
print(A ^ B)
print(A.symmetric_difference(B))
```

# Set Checks

```python
A = {1, 2}
B = {1, 2, 3, 4}

# issubset()
print(A.issubset(B))
```

```python
A = {1, 2, 3, 4}
B = {1, 2}

# issuperset()
print(A.issuperset(B))
```

```python
A = {1, 2}
B = {3, 4}

# isdisjoint()
print(A.isdisjoint(B))
```
