# Dictionary

## Initialization
- Key-value pair collection
- Mutable
- Keys are unique

```python
# Empty dictionary
d = {}

print(d)
```

```python
# Dictionary with values
d = {
    "name": "Keerthi",
    "age": 21
}

print(d)
```

# defaultdict

```python
from collections import defaultdict

# int -> default 0
d = defaultdict(int)

print(d["a"])
```

```python
from collections import defaultdict

# list -> default []
d = defaultdict(list)

print(d["a"])
```

```python
from collections import defaultdict

# set -> default set()
d = defaultdict(set)

print(d["a"])
```

# Insert Operations

```python
d = {}

# Single insert
d["a"] = 1

print(d)
```

```python
d = {"a": 1}

# Multiple inserts
d.update({
    "b": 2,
    "c": 3
})

print(d)
```

# Update Operations

```python
d = {
    "a": 1,
    "b": 2
}

# Update value
d["a"] = 100

print(d)
```

```python
d = {
    "a": 1
}

# Multiple updates
d.update({
    "b": 2,
    "c": 3
})

print(d)
```

# Delete Operations

```python
d = {
    "a": 1,
    "b": 2
}

# pop()
value = d.pop("a")

print(value)
print(d)
```

```python
d = {
    "a": 1,
    "b": 2
}

# popitem()
item = d.popitem()

print(item)
print(d)
```

```python
d = {
    "a": 1,
    "b": 2
}

# clear()
d.clear()

print(d)
```

```python
d = {
    "a": 1,
    "b": 2,
    "c": 3
}

# Dictionary comprehension
new_d = {k:v for k, v in d.items() if k != "a"}

print(new_d)
```

# Accessing Elements

```python
d = {
    "name": "Keerthi",
    "age": [21,22,23]
}

# Access using key
print(d["name"])
print(d["age"])
```

```python
d = {
    "a": 1,
    "b": 2,
    "c": 3
}

# keys()
for key in d.keys():
    print(key)
```

```python
d = {
    "a": 1,
    "b": 2,
    "c": 3
}

# values()
for value in d.values():
    print(value)
```

```python
d = {
    "a": 1,
    "b": 2
}

# items()
for key, value in d.items():
    print(key, value)
```

# Searching

```python
d = {
    "a": 1,
    "b": 2
}

# Key search
print("a" in d)
print("z" in d)
```

```python
d = {
    "a": 1,
    "b": 2
}

# Value search
print(2 in d.values())
print(10 in d.values())
```
