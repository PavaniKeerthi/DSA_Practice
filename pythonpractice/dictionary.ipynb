```python
d = {}

print(d)
```

```python
from collections import defaultdict

d = defaultdict(int)

print(d["a"])
```

```python
d = {}

d["a"] = 1

d.update({"b": 2, "c": 3})

print(d)
```

```python
d = {"a": 1, "b": 2}

x = d.pop("a")

print(x)
print(d)
```

```python
d = {"a": 1, "b": 2}

print(d.popitem())
print(d)
```

```python
d = {"a": 1, "b": 2}

d.clear()

print(d)
```

```python
d = {"a": 1, "b": 2}

for key in d.keys():
    print(key)
```

```python
d = {"a": 1, "b": 2}

for value in d.values():
    print(value)
```

```python
d = {"a": 1, "b": 2}

for key, value in d.items():
    print(key, value)
```

```python
d = {"a": 1, "b": 2}

print("a" in d)
```

```python
d = {"a": 1, "b": 2}

print(2 in d.values())
```
