```python
s = set()

print(s)
```

```python
s = {1, 2}

s.add(3)

print(s)
```

```python
s = {1, 2}

s.update([3, 4])

print(s)
```

```python
s = {1, 2, 3}

new_set = s - {2}

print(new_set)
```

```python
s = {10, 20, 30}

x = s.pop()

print(x)
print(s)
```

```python
s = {1, 2, 3}

s.clear()

print(s)
```

```python
numbers = [1, 2, 3, 4]

squares = {x**2 for x in numbers}

print(squares)
```

```python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)
print(A & B)
print(A - B)
print(A ^ B)
```

```python
A = {1, 2}
B = {1, 2, 3}

print(A.issubset(B))
print(B.issuperset(A))
print(A.isdisjoint({4, 5}))
```
